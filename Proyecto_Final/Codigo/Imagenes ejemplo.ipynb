```python
import os 
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
from random import choice
import cv2
```

```python
root_dir = "../Dataset"
```

```python
# Create dictionary that maps integer class index to string with name of class
file = os.path.join(root_dir, "classes.txt")   # txt file containing the name of the corresponding class

with open(file, "r") as file:
    lines = file.readlines()

classes = {}

# Get key and value for each class (line of txt file)
for line in lines:
    key, value = line.strip().split(" ", 1)
    classes[int(key.strip()) - 1] = value.strip()
```

```python
# Elegir 16 clases aleatorias para mostrar
idxs = np.random.randint(102, size=16)

classes_rand = [classes[x] for x in idxs]
```

```python
fig = plt.figure(figsize=(8,8))

for i, clase in enumerate(classes_rand):
    ax = plt.subplot(4, 4, i+1)
    path_imgs = os.path.join("train", clase)
    all_imgs = os.path.join(path_imgs, "*")
    img = cv2.imread(choice(glob(os.path.join(root_dir, all_imgs))))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    ax.imshow(img)
    ax.set_title(clase, fontsize=8)
    ax.axis("off")
plt.show()
plt.tight_layout()
```
